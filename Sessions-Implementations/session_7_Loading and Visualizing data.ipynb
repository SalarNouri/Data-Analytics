##### What to do with NaN? we decide according to the data and limitations we are working on:

1. Through them away(index or columns?)
2. fill them (different methods: 0 | mean | back | forward)
3. interpolation



###### if u have enough data (only 5% is NaN) no matter if u eliminate them but if not, dont do so:)

```python
# First eliminate some
electricity=electricity[range(1971,2015)]
oil=oil[range(1971,2015)]
```

```python
# then fill the rest
electricity.fillna(value=0)
```

```python
electricity.fillna(method='ffill')
# fill the empty arrays by the upper one
```

```python
electricity.fillna(method='ffill',axis=1)
# to fill according to each country
# but note that NaN s still stay Nan! it's because of "ffill" which means FORWARD fill
```

```python
electricity.fillna(method='bfill',axis=1)
# thats what we wanted!!
```

```python
# SO
electricity=electricity.fillna(method='bfill',axis=1)
oil=oil.fillna(method='bfill',axis=1)
```

```python
# let's check again
electricity.isna().sum()
oil.isna().sum()
```

# Rafactor your code:

```python
oil = pd.read_csv('oil_consumption_per_cap.csv',index_col='country')
oil.columns=[int(s) for s in oil.columns]

oil=oil[range(1971,2015)]

oil=oil.fillna(method='bfill',axis=1)

display(oil)
print(oil.index)
print(oil.columns)
```

```python
electricity = pd.read_csv('electricity_use_per_person.csv',index_col='country')
electricity.columns=[int(s) for s in electricity.columns]
electricity=electricity.loc[oil.index,:]

electricity=electricity[range(1971,2015)]

electricity=electricity.fillna(method='bfill',axis=1)

display(electricity)
print(electricity.index)
print(electricity.columns)
```
