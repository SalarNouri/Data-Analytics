# Pivot table

### میانگین ستون ها را برای اندیس های مطلوب مشخص میکنه

```python
pd.pivot_table(df , index = ['race'] , aggfunc = 'mean')
# پس با ایندکس مشخص میکنیم کل دیتا رو بر اساس چه دسته بندی ای بررسی کنه، بعدش با متغیر بعدی که بهش میدیم مشخص میشه که 
# برای اون دسته ها بره چی رو بر حسب ستون ها (که به نوعی همون فیچرهامون هستن) حساب کنه
```

```python
pd.pivot_table(df , index = ['race'] , aggfunc = 'mean').hours
```

```python
pd.pivot_table(df , index = ['race'] , columns = ['income'] , aggfunc = 'mean').hours
```

```python
pd.pivot_table(df , index = ['race'] , columns = ['income'] , aggfunc = 'mean')
```

```python
pd.pivot_table(df , index = ['race'] , columns = ['income'] ,values = ['hours'], aggfunc = 'mean')
```

```python
pd.pivot_table(df , index = ['race'] , columns = ['income','gender'] ,values = ['hours'], aggfunc = 'mean')
```

```python
pd.pivot_table(df , index = ['gender','income'] , aggfunc = 'mean')
```

# Joining tables

```python
df1 = pd.DataFrame({'A': ['A11', 'A12'],
                    'B': ['B11', 'B12'],
                    'C': ['C11', 'C12'],
                    'D': ['D11', 'D12']},
                     index=[0,1])
df2 = pd.DataFrame({'A': ['A1', 'A2', 'A3'],
                    'B': ['B1', 'B2', 'B3'],
                    'C': ['C1', 'C2', 'C3'],
                    'D': ['D1', 'D2', 'D3']},
                     index=[1,2,3])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1 , df2])
```

```python
pd.concat([df1 , df2], axis = 1, join = 'inner')
# ایندکس مشترک رو کنار هم قرار داده چون کلمه inner میگه این کارو کن
```

```python
pd.concat([df1 , df2], axis = 1, join = 'outer')
```

# merge
برای اتصال دیتا فریم ها یا سری ها بکار میره

```python
df1 = pd.DataFrame({'id': [81010, 81011,81012,81013],
                    'DA': [20, 15,18.75,10],
                    'SI': [16, 17,20,13],
                    'IL': [17,13,14,11]})
df2 = pd.DataFrame({'id': [81011, 81012,81013,81014,81015],
                    'GPA': [16.5,18,15,19.3,17.7],
                    'uni': ['A0','A1','A2','A3','A5']})
```

```python
df1
```

```python
df2
```

#### اینبار میخواهیم براساس شماره دانشجویی ها جدول هارا یکی کنیم.
برای همین نمیتونیم از ایندکس استفاده کنیم

```python
pd.merge(df1,df2, on = 'id')
# only common data has returned
```

```python
pd.merge(df1,df2, on = 'id' , how = 'outer') # outer or inner
#که دیفالتش رو اینر هست
# اجتماع داده ها رو نشون داده
```

```python
pd.merge(df1,df2, on = 'id' , how = 'right') # براساس دیتا دوم میاره
# left
```

```python
pd.merge(df1,df2, on = 'id' , how = 'left')
```

# Join
براساس ایندکس میاد دیتافریم هارا مرتبط میکند.

```python
df1.join(df2 , lsuffix= '1111' , rsuffix= '222')
# فقط براساس "ایندکس" میاد دیتا رو مرتبط میکنه
# lsuffix and rsuffix :
# اگر در این دو دیتاست، ستون هایی با اسم های تکراری داشته باشیم بهشون پسوند میده تا مشخص بشه از کدوم دیتاست هستند
```

```python
df1.set_index('id').join(df2 , lsuffix= '1' , rsuffix= '2')
```

```python
df1.join(df2.set_index('id') , lsuffix= '1' , rsuffix= '2')
```

```python
df1.set_index('id').join(df2.set_index('id') , lsuffix= '1' , rsuffix= '2')
```
