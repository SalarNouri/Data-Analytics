# Replace unrelated values

```python
df = pd.read_csv('team.csv')
df
```

```python
# همانطور که در دیتای بالا میبینی، برخی داده ها هستن که به عنوان "نن" شناخته نشدن
# پس ما دو راه داریم: اولی اینه که ریپلیس کنیم شون
```

```python
df.replace({'?':np.nan, '_':np.nan}) # حالا برای اینکه تغییرات روی داده اعمال بشه از 
# inplace استفاده میکنیم
```

```python
df.replace({'?':np.nan, '_':np.nan}, inplace = True)
```

### اما راه دوم: همون اول مشخص کنیم موقع فراخوانی داده ها چی رو میسینگ ول بگیره

```python
missing_values = ['?','_']
df = pd.read_csv('team.csv', na_values = missing_values)
df
```

## Handelling missing values

### there are several methods for handelling missing vals and we talk abt some:
    
    a.Removing data: when we have a large data samples and the amount of NaN s is less enough.
    

```python
df.dropna(axis = 0, how = 'all') # سطری که همه مقادیرش نال باشن
```

```python
df.dropna(axis = 1, how = 'all',inplace = True)# ستونی که همه مقادیریش نال بودن میحذفه
```

```python
df.dropna(axis = 0, how = 'any')# هرکی نال داره رو "سطرشو" میحذفه
```

```python
df.dropna(axis = 0, thresh = 2)# سطرهایی که حداقل دو مقدار غیرنال داشته باشن رو نگه میداره
```

### اما این روشی که گفته شد بیشتر برای حذف سطرها بکار میره و برای حذف یه فیچر یا ستون قاعده ی دیگری وجود داره که سرانگشتی هست،
اینکه اگر تعداد میس ولیو ها از 60تا 70 درصد کل داده بیشتر باشه اون فیچرو بحذفیم( واضحه که این درصد که گفته شد بستگی به موضوع و سلیقه تحلیلگر داره)

```python
df.salary.fillna(df.salary.mean())
# but if you have outliers, this method is not recommended and 
# its better to use "median"!
# ++ its easy
# ++ for small datasets

#-- doesn't have high accuracy
#-- unable to be used in categorical datasets
# همچنین اگر اوتلایر داشته باشیم پاسخ رو خراب میکنه
```
