# داده تکراری
    اول ببینیم آیا داده تکراری وجود داره یا نه؟

### انواع روش های نشان دادن تکرارها

```python
df.duplicated(subset = ['cases','deaths'],keep = False)
# subset ===> میگه بر اساس کدوم ستون ها بررسی کنم تکرار رو؟
# نتیجه:
# اونایی که فالس هستن ینی تکراری نبودن و بالعکس
# اونایی ک ترو هستن ینی تکرار شدن( البته توجه داری که بر اساس اون ستون های مشخص شده هستش)
```

```python
# اگر بخوایم دقیقا مقدار دیتافریم رو در مقادیر تکراری ببینیم به صورت زیر بدست میاد

df.loc[df.duplicated(subset = ['cases','deaths'],keep = False)]
```

```python
# first اولین مقدار تکراری رو نگه میداره و بقیه رو بیرون میریزه
df.loc[df.duplicated(subset = ['cases','deaths'],keep = 'first')]
```

```python
# last آخری رو نگه میداره
df.loc[df.duplicated(subset = ['cases','deaths'],keep = 'last')]
```

### حذف تکرارها

```python
# سطر هایی که "همه" ی مقادیر شون یکی هست رو نشون بده
df.loc[df.duplicated(keep='first')]
# حالا میخوایم این رو حذف کنیم
# روش اول:
#df.drop(df.loc[df.duplicated(keep='first')].index)

# روش دوم:
# --------------به دیکته ش توجه کن!!-------------------
df.drop_duplicates(keep = 'first' , inplace = True, ignore_index = True)
# این دستور برای این هست  که توی شماره ی سطرها منظور همون ایندکس فاصله نیوفته ignor_index
```

```python
# تاریخ میلادی رو به شمسی تبدیل کنیم
from persiantools.jdatetime import JalaliDate
```

```python
JalaliDate.to_Jalali(2020,11,13)
# فقط روی یک المان میتونه اعمال بشه و ما نمیتونیم روی سری یا دیتافریم اعمالش کنیم
```

```python
def my_function (in_str):
    year = int(in_str[:4]) # از اولی تا چهارمی
    month = int(in_str[5:7]) # از 5 تا7
    day = int(in_str[7:])
    return JalaliDate.to_Jalali(year,month,day)
```

```python
df.dateRep.apply(my_function)
```

# بیان تابع به صورت کوتاه تر

Lambda

```python
df.dateRep.apply(lambda x: JalaliDate.to_Jalali(int(x[:4]),int(x[5:7]),int(x[7:])))
```
