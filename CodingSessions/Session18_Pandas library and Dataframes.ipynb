# Selection by position
چون در دیتا ایران که از دیتاست اصلی جدا کردیم ایندکس ها را ریست نکردیم، شماره شون از صفر شروع نشده. برای همین گاهی لازم میشه مثل این شرایط، که ایندکس را نه براساس نامی که داره، بلکه براساس موقعیت مکانی ش هم دسترسی داشته باشیم.

```python
# boolean indexing
df_ir.loc[df_ir.cases > 3000]
```

# Where method
در دستور بولین قبلی، ابعاد دیتاست تغییر میکنه، یعنی فقط سطرها یا ستون هایی که شرط در اونها برقرار هست رو میده. ولی با دستور "ور" همه ی دیتاست رو میده با این تفاوت که اونهایی که در شرط صدق نمیکنن رو "نن" نشون میده.

```python
df_ir.where(df_ir.cases > 3000)

df_ir.where(df_ir.cases > 3000,0)# مقدار 0 برگردونه بجای اونا
```

# Add a new column
# دو روش برای اینکار وجود داره


```python
df_ir['ratio'] = df_ir.deaths/df_ir.cases
df_ir
```

## روش دوم

```python
df_ir.insert(1,'ratio_percent',df_ir.ratio*100)
df_ir.head()
```

# Delete a column

```python
df_ir.drop(['ratio_percent','ratio'],axis=1, inplace = True)
df_ir
```

# تغییر نام ستون

```python
df_ir.rename(columns = {'Cumulative_number_for_14_days_of_COVID-19_cases_per_100000':'CNC'},inplace = True)
df_ir.head(2)
```
