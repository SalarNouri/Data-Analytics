```python
import numpy as np
from scipy.signal import argrelextrema

# ==================================== For part (A1)
# -- Problem(a): Determine the indices of the local maxima
np.random.seed(0)
x = np.random.randint(10, size=(1, 20) )
max_idx = argrelextrema(x[0, :], np.greater)
print("Local maximum indexes are:", max_idx)

# -- Problem(b): Find top five maximum indexes
N = 5 
max_selected_idx = np.argpartition(x, -N)[-N:]
print("Top five maximum indexes are:", max_selected_idx)

# -- Problem(c): Find common values in two array
similar_values = np.intersect1d(x[0, max_idx], x[0, max_selected_idx])
print("Similar values of arrays:", similar_values)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ==================================== For part (A3)
# -- Problem(a): Loading and displying the dataset
filename = 'cars.csv'
df_cars = pd.read_csv(filename)
df_cars.drop_duplicates(keep = 'last', inplace = True)
display(df_cars)

# -- Problem(b): Finding Make vaules and plot Make values
NUM_OF_DISPLAYED_MAKE = 10
result_make = []
for item in df_cars.Make.unique():
    df_make = df_cars[df_cars['Make'] == item]
    result_make.append(
            {
                "Make": item,
                "value": len(df_make['Make']) 
            }
        )
NUM_OF_DISPLAYED_MAKE = 10
result_make = pd.DataFrame(result_make).sort_values(by='value', ascending=False)
selected_make = result_make.iloc[0:NUM_OF_DISPLAYED_MAKE]
plt.figure()
selected_make.plot.bar(x='Make', y='value', color='blue')
plt.grid()
plt.title("The cars' top 10 Make company")

# Plot scatterplot
plt.figure()
plt.scatter(df_cars['Engine HP'], df_cars['MSRP'], color='blue')
plt.grid()
plt.xlabel('Cars Enging HP')
plt.ylabel('Cars MSRP')
plt.title('Engine HP correaltion with MSRP');
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ==================================== For part (B3)
# -- Problem(a): Loading and displying the dataset
filename = 'cars.csv'
df_cars = pd.read_csv(filename)
df_cars.drop_duplicates(keep = 'last', inplace = True)
display(df_cars)

# -- Problem(b): Finding Vehicle Size vaules and plot Make values
result_vehicle = []
for item in df_cars['Vehicle Size'].unique():
    df_vehicle = df_cars[df_cars['Vehicle Size'] == item]
    result_vehicle.append(
            {
                "Vehicle Size": item,
                "value": len(df_vehicle['Vehicle Size']) 
            }
        )
result_vehicle = pd.DataFrame(result_vehicle).sort_values(by='value', ascending=False)
plt.figure()
result_vehicle.plot.bar(x='Vehicle Size', y='value', color='blue')
plt.grid()
plt.title("The Num of cars in each Vehicle Size")

# -- Problem(c): 
df_record = []
for item in df_cars['Vehicle Style'].unique():
    df_vehicle_style = df_cars[df_cars['Vehicle Style'] == item]
    for itm in df_cars['Vehicle Size'].unique():
        df_vehicle_size = df_vehicle_style[df_vehicle_style['Vehicle Size'] == itm]
        df_record.append(
            {
                'Vehicle Style': item,
                'Vehicle Size': itm,
                'Record': len(df_vehicle_size)
            }
        )
df_record = pd.DataFrame(df_record)
pvt = pd.pivot_table(df_record, values='Record', index=['Vehicle Style'],
                    columns=['Vehicle Size'], aggfunc=np.sum, fill_value=0)
print('The pivot table of Correspond values:')
display(pvt)
pvt.columns = list(pvt.columns)
pvt.reset_index(inplace=True)
result = pvt.copy()
print("The Correspond values of Vehicle Style and Vehicle Size")
display(result)
```
